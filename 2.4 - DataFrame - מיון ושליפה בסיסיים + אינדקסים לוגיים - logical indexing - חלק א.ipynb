2.4 - DataFrame - מיון ושליפה בסיסיים + אינדקסים לוגיים - logical indexing - חלק א

```python
import pandas as pd
```

```python
url = 'https://s3.eu-west-1.amazonaws.com/data.cyber.org.il/virtual_courses/introdata/colab/artists.csv'
artist = pd.read_csv(url, sep=None, engine='python')
```

SORT_VALUES מיון טבלה על פי 

```python
artist_by_popularity = artist.sort_values(['popularity','followers'], ascending = False)
```

```python
artist_by_popularity.head()
```

```python
artist_by_popularity.tail()
```

שליפת נתונים מהטבלה 

```python
artist_by_popularity.iloc[10:20,:]
```

```python
artist_by_popularity[['name','popularity','followers']]
```

```python
artist_by_popularity.loc[:,['name','popularity','followers']]
```
